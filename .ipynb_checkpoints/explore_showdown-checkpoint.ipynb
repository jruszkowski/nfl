```python
#!usr/bin/python3

import sys
import projections
import pandas as pd
from itertools import combinations
import numpy as np
from joblib import Parallel, delayed
import datetime
from collections import defaultdict

def remove_suffix(p, fanduel=False):

    if fanduel:
        p = p.replace(' Jr.', '')
        if fanduel_d.get(p.strip()):
            p = fanduel_d.get(p.strip())
    if (not fanduel) and defense.get(p.strip()):
        p = defense.get(p.strip())
    p = p.replace(' III', '')
    p = p.replace(' II', '')
    p = p.replace(' IV', '')
#     if p != 'Las Vegas':
#         p = p.replace(' V', '')
    p = p.replace('AJ ', 'A.J. ')
    p = p.replace('DJ Chark Jr.', 'D.J. Chark')
    p = p.replace('Marvin Jones Jr.', 'Marvin Jones')
    p = p.replace('Robert Griffin', 'Robert Griffin III')
    p = p.replace('DJ ', 'D.J. ')

    return p

defense = {
     'Steelers': 'Pittsburgh',
     'Chargers': 'Los Angeles Chargers',
     'Browns': 'Cleveland',
     'Vikings': 'Minnesota',
     'WAS Football Team': 'Washington',
     'Dolphins': 'Miami',
     'Lions': 'Detroit',
     'Patriots': 'New England',
     'Ravens': 'Baltimore',
     'Buccaneers': 'Tampa Bay',
     'Saints': 'New Orleans',
     'Chiefs': 'Kansas City',
     'Bengals': 'Cincinnati',
     'Texans': 'Houston',
     'Broncos': 'Denver',
     'Eagles': 'Philadelphia',
     'Panthers': 'Carolina',
     'Cardinals': 'Arizona',
     'Packers': 'Green Bay',
     'Titans': 'Tennessee',
     'Cowboys': 'Dallas',
     'Rams': 'Los Angeles Rams',
     'Seahawks': 'Seattle',
     'Jaguars': 'Jacksonville',
     'Falcons': 'Atlanta',
     'Jets': 'New York Jets',
     'Colts': 'Indianapolis',
     'Raiders': 'Las Vegas'
 }

fanduel_d = {
    'Pittsburgh Steelers': 'Pittsburgh',
    'Miami Dolphins': 'Miami',
    'Baltimore Ravens': 'Baltimore',
    'Cleveland Browns': 'Cleveland',
    'Los Angeles Chargers': 'Los Angeles Chargers',
    'Minnesota Vikings': 'Minnesota',
    'New England Patriots': 'New England',
    'Green Bay Packers': 'Green Bay',
    'New Orleans Saints': 'New Orleans',
    'Washington Football Team': 'Washington',
    'Indianapolis Colts': 'Indianapolis',
    'Philadelphia Eagles': 'Philadelphia',
    'Denver Broncos': 'Denver',
    'Tennessee Titans': 'Tennessee',
    'Cincinnati Bengals': 'Cincinnati',
    'Houston Texans': 'Houston',
    'Detroit Lions': 'Detroit',
    'Carolina Panthers': 'Carolina',
    'Jacksonville Jaguars': 'Jacksonville',
    'Atlanta Falcons': 'Atlanta',
    'New York Jets': 'New York Jets',
    'Dallas Cowboys': 'Dallas',
}

plyr_dict = projections.get_nf_projections()
plyr_dict.update(projections.get_nf_d_projections())
```

```python
defense = {'d': 'DST', 'f': 'D'}
plyr_names = {'d': 'Name', 'f': 'Nickname'}
salary = {'d': 50000, 'f': 60000}
```

```python
f = 'dk_showdown.csv'
df = pd.read_csv('inputs/' + f)
players = df['Name'].to_list()
players = {p: remove_suffix(p) for p in players}
df['Name'] = df['Name'].replace(players)
df = df.drop(df.loc[(df['Name'] == 'Josh Johnson') & (df.TeamAbbrev == 'NYJ')].index)
df = df.drop(df.loc[(df['Name'] == 'Ryan Griffin') & (df.TeamAbbrev == 'TB')].index)
df_cpt = df.loc[df['Roster Position']=='CPT']
df_flex = df.loc[df['Roster Position']=='FLEX']
```

```python
f = 'fd_showdown.csv'
df = pd.read_csv('inputs/' + f)
df = df.drop(df.loc[(df['Nickname'] == 'Ryan Griffin') & (df.Team == 'TB')].index)
assert len(df) == len(df['Nickname'].unique())
players = df['Nickname'].to_list()
players = {p: remove_suffix(p, True) for p in players}
df['Nickname'] = df['Nickname'].replace(players)
df = df.set_index('Nickname')
df['Projection'] = pd.DataFrame.from_dict(plyr_dict, orient='index')
df = df.loc[df['Projection'] > 1]
df = df.reset_index()
df_cpt = df.copy()
df_flex = df.copy()

df_cpt.head()
```

```python
df_cpt = df_cpt.set_index(plyr_names[f[0]])
df_cpt['Projection'] = pd.DataFrame.from_dict(plyr_dict, orient='index') * 1.5
df_cpt = df_cpt.reset_index()
df_cpt[plyr_names[f[0]]] = df_cpt[plyr_names[f[0]]].apply(lambda x: x.strip())
assert len(df_cpt[plyr_names[f[0]]]) == len(df_cpt[plyr_names[f[0]]].unique())
df_cpt.set_index([plyr_names[f[0]]], inplace=True)
df_cpt = df_cpt[df_cpt['Projection'] > 1]
df_cpt = df_cpt.reset_index()
df_pos_salary = df_cpt.groupby(['Salary'])['Projection'].agg([np.mean, np.max])
df_pos_salary = df_pos_salary.reset_index()
df_cpt = df_cpt.merge(df_pos_salary, on=['Salary'])
df_cpt = df_cpt[df_cpt['Projection'] == df_cpt['amax']]
df_cpt.set_index(plyr_names[f[0]], inplace=True)
df_cpt.head()
```

```python
df_flex = df_flex.set_index(plyr_names[f[0]])
df_flex['Projection'] = pd.DataFrame.from_dict(plyr_dict, orient='index')
df_flex = df_flex.reset_index()
df_flex[plyr_names[f[0]]] = df_flex[plyr_names[f[0]]].apply(lambda x: x.strip())
assert len(df_flex[plyr_names[f[0]]]) == len(df_flex[plyr_names[f[0]]].unique())
df_flex.set_index([plyr_names[f[0]]], inplace=True)
df_flex = df_flex[df_flex['Projection'] > 1]
df_flex = df_flex.reset_index()
df_pos_salary = df_flex.groupby(['Salary'])['Projection'].agg([np.mean, np.max])
df_pos_salary = df_pos_salary.reset_index()
df_flex = df_flex.merge(df_pos_salary, on=['Salary'])
df_flex = df_flex[(df_flex['Projection'] >= df_flex['mean'])]
df_flex.set_index(plyr_names[f[0]], inplace=True)
df_flex = df_flex.reset_index()
df_pos = df_flex.groupby(['Projection'])['Salary'].agg([np.min])['amin']
df_pos = df_pos.reset_index()
df_flex = df_flex.merge(df_pos, on=['Projection'])
df_flex = df_flex[(df_flex['Salary'] <= df_flex['amin'])]
df_flex.set_index(plyr_names[f[0]], inplace=True)
df_flex.head()
```

```python
flex_plyr_list = df_flex.to_dict(orient='index')
cpt_plyr_list = df_cpt.to_dict(orient='index')
flex_salary = {k: v['Salary'] for k,v in flex_plyr_list.items()}
cpt_salary = {k: v['Salary'] for k,v in cpt_plyr_list.items()}
flex_proj = {k: v['Projection'] for k,v in flex_plyr_list.items()}
cpt_proj = {k: v['Projection'] for k,v in cpt_plyr_list.items()}

def return_combos(plyr_list, count):
    return list(combinations(plyr_list, count))

flex_ct = {'d': 5, 'f': 4}

flex = return_combos(flex_plyr_list, flex_ct[f[0]])

total_list = []
for c in cpt_plyr_list.keys():
    for x in flex:
        if c not in x:
            total_list.append([c, *x])

df_total = pd.DataFrame(total_list)            
df_salary_flex = df_total.iloc[:,1:].replace(flex_salary)
df_salary_cpt = df_total.iloc[:,0].replace(cpt_salary)
df_proj_flex = df_total.iloc[:,1:].replace(flex_proj)
df_proj_cpt = df_total.iloc[:,0].replace(cpt_proj)
total_salary = df_salary_flex.sum(axis=1) + df_salary_cpt
total_projections = df_proj_flex.sum(axis=1) + df_proj_cpt
df_total['salary'] = total_salary
df_total['projection'] = total_projections
df_total = df_total.loc[df_total['salary'] <= salary[f[0]]]
df_total.sort_values('projection', ascending=False).head(10)
```
